## scope of variables
## local scope-variables inside a function


```python
def my_function():
    local_var=10
    print(f"inside function:{local_var}")
```

```python
my_function()
```

```python
## should be errror in inside functione;
print(local_var)
```

## global scope
## variable defined outside function
## globle variable can access the both inside and outside function

```python
global_var=100
def my_function():
    print(f"outside function:{global_var}")
```

```python
my_function()
```

```python
## should not be error
print(global_var)
```

```python
local_var
```

```python
amount=100

def my_amount(amount):
    amount+=1
    print(f"my amount:",amount)
```

```python
total=200

def my_total():
    global total
    total+=1
    print(f"my total:",total)
```

```python
my_total()
```

```python
my_total()
my_total()
print(f"count outside:{total}")
```
